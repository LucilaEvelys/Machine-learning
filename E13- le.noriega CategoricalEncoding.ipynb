```python
# ParticiÃ³n del x_train y el x_test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

Se define un rango de profundidad para hacer la prueba de 10.

```python
# list of values to try for max_depth
max_depth_range = range(1, 10)

# list to store the average RMSE for each value of max_depth
accuracy_scores = []

# use 10-fold cross-validation with each value of max_depth
from sklearn.model_selection import cross_val_score
#from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

for depth in max_depth_range:
    clf = RandomForestClassifier(max_depth=depth, random_state=1)
    accuracy_scores.append(cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy').mean())
```

```python
accuracy_scores
```

Con el cÃ¡lculo de las precisiones, se establece que la profundidad que ofrece una mayor precisiÃ³n es igual a 7. Con esa profundida se procede a la selecciÃ³n de los parÃ¡metros para el random forests (entrenamiento y predicciÃ³n).

```python
clf = RandomForestClassifier(max_depth=7, random_state=1)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
print ('accuracy:',metrics.accuracy_score(y_pred, y_test))
```

Con una profundidad de 7, la precisiÃ³n del modelo de random forests es igual a 0,95.

### PCA

```python
from sklearn.decomposition import PCA
```

```python
X=df.drop(['price'], axis=1)
X.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_, y, test_size=0.33, random_state=42)
```

```python
clf = RandomForestClassifier(max_depth=7, random_state=1)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred1=clf.predict(X_test)
```

```python
print ('accuracy:',metrics.accuracy_score(y_pred1, y_test))
```

Con el mÃ©todo de encode PCA, se obtiene un accuracy de 0,91. Respecto al modelo anterior disminuye en aproximadamente un punto. Esta es la evidencia de que en la medida en que el modelo se va haciendo mÃ¡s manejable con la inclusiÃ³n de las dummies se castiga la eficiencia del modelo.

## Binary

Binary encoding for categorical variables, similar to onehot, but stores categories as binary bitstrings.

```python
import category_encoders as ce
```

```python
# crear la representaciÃ²n en binario del numero de cada categoria (bit)
X_ = ce.BinaryEncoder().fit_transform(df.drop(['price'], axis=1))
```

```python
X_.head()
```

```python
pd.Series(cross_val_score(RandomForestClassifier(n_estimators=10), X_, y, cv=10, scoring='accuracy')).describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_, y, test_size=0.33, random_state=42)
```

```python
clf = RandomForestClassifier(max_depth=7, random_state=1)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred1=clf.predict(X_test)
```

```python
print ('accuracy:',metrics.accuracy_score(y_pred1, y_test))
```

Con mÃ©todo de encoding evaluados en este Ã­tem se aumenta en aproximadente tres puntos respecto al modelo usado en el Ã­tem 13.3 en accuracy. Con los valores obtenidos se concluye que el mÃ©todo Binary ofrece mejor desempeÃ±o que el mÃ©todo PCA.
