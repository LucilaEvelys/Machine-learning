```python
def similarity(data):
    from difflib import SequenceMatcher as SM
    results=pd.DataFrame(0, index=data.index, columns=data.index)
    for i in results.index:    
        for j in results.index:
            if pd.isna(data.Text[i])==1 or pd.isna(data.Text[j])==1:
                results.loc[j,i]=0            
            else:
                results.loc[j,i]=SM(None,data.Text[i],data.Text[j]).ratio()
                
                    
            globals()['Tx%s' % i] = data.loc[i, 'Text']
            #results.loc[j,i]=SM(None,data.Text[i],data.Text[j]).ratio()
            #print(SM(None,data.Text[i],data.Text[j]).ratio())
    
    #assert results.shape == (data.shape[0], data.shape[0])
    return results
```
